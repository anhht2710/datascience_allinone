```python
def newton_method(N, num_loops):
    '''
    This function aims to compute square root for the number N
    
    N -- the number needs to take the square root
    num_loops -- number of loops used for this optimization
    '''
    
    x_n = N/2
    
    for i in range(num_loops):
        x_np1 = (x_n + N/x_n) / 2
        x_n = x_np1
        
    return x_np1

print(newton_method(4, 100))
print(newton_method(2, 100))
```

```python
# Compute the circumference of a unit circle
import math

def compute_length(degree):
    sin = math.sin(math.radians(degree))
    length = sin*(180/degree)

    return length

print(compute_length(240))
```

```python
# Compute the area of a unit circle
import math

def compute_area(degree):
    sin_degree = math.sin(math.radians(degree))
    cos_degree = math.cos(math.radians(degree))

    area_degree = sin_degree * cos_degree / 2
    area = area_degree*(360/degree)

    return area

print(compute_area(90))
```

```python
# Compute the area of a unit circle
import numpy as np

def get_y(x):
    y = np.sqrt(1 - x*x)
    return y

# set delta_x
delta_x = 0.0001

# get the value of x
segments = np.arange(-1, 1+delta_x, delta_x)

# compute rectangle widths
begin = segments[0:-1]
end = segments[1:]
width = end - begin

# compute heights
y = get_y(begin)

# compute area
area = np.sum(width*y)
print(area*2)
```

## 5. Conditions <a name="part5"></a>

If-else Condition and Math \
Input: `a` and `b` \
if `a` = 0 then y = $b^2$ \
if `a` = `K` then y = $\sqrt{b}$


```python
import math

def function_1(a,b):
    # a constant
    K = 10

    result = 0
    if a==0:
        result = b*b
    elif a==K:
        result = math.sqrt(b)

    return result

print(function_1(0, 4))
print(function_1(10,4))
```

Input: `a` and `b` \
$
y = \frac{a}{K} \left(\sqrt{b}\right) + \frac{K - a}{K} b^2
$


```python
import math
def function_2(a,b):
    # a constant
    K = 10

    result = (a/K)*math.sqrt(b) + ((K-a)/K)*b*b

    return result

print(function_2(0, 4))
print(function_2(10, 4))
```

Input: `a` and `b` \
if `a` > 0 then `y` = y = $b^2$ \
if `a` < 0 then y = $\sqrt{b}$

```python
import math

def function_3(a,b):
    result = 0

    if a>0:
        result = b*b
    elif a<0:
        result = math.sqrt(b)

    return result

print(function_3(2, 4))
print(function_3(-2, 4))
```

Input `a` and `b`then \
$y = \frac{b^2 + \sqrt{b}}{2} + \frac{b^2 - \sqrt{b}}{2} \frac{|a|}{a}$


```python
import math

def function_4(a,b):
    result = (b*b + math.sqrt(b))/2 + (b*b - math.sqrt(b))/2 * (abs(a)/a)
    return result

print(function_4(2, 4))
print(function_4(-2, 4))
```

Input `a` and `b`

if `a` = 0 then `y` = $b^2$  
if `a` = 1 then `y` = $\sqrt{b}$  
if `a` = 2 then `y` = `b`

```python
import math

def function_5(a,b):
    result = 0
    
    if a==0: result = b*b
    elif a==1: result = math.sqrt(b)
    elif a==2: result = b

    return result

print(function_5(0, 4))
print(function_5(1, 4))
print(function_5(2, 4))
```

```python
import math
import numpy as np

def function_6(a,b):
    vector_v = np.array([0, 0, 0])
    vector_v[a] = 1

    vector_u = np.array([b*b, math.sqrt(b), b])

    result = np.dot(vector_v, vector_u)
    return result

print(function_6(0, 4))
print(function_6(1, 4))
print(function_6(2, 4))
```

## 6. Overflow and Underflow <a name="part6"></a>

* **Overflow**: Happens when a calculation exceeds the limits of a data type. In Python, this doesn't affect integers, but floating-point numbers can result in `inf`.

* **Underflow**: Occurs when a number is too small to be represented and is rounded down to zero, typically affecting floating-point numbers.

```python
result = 1e-100
print(result)
```

```python
result = 1e-1000
print(result)
```

```python
result = 1e100
print(result)

x = 10 ** 100  
print(x)
```

```python
result = 1e1000
print(result)
```

```python
import math
x = math.exp(1000)  
print(x)  
```

The error occurs because the value of `math.exp(1000)` is too large for Python to handle, \
resulting in an OverflowError. The exponential function grows very quickly,   \
and `exp(1000)` exceeds the maximum limit for floating-point numbers in Python.

**Softmax function**

$f(x_i) = \frac{e^{x_i}}{\sum_j e^{x_j}}$

$0 \leq f(x_i) \leq 1$

$\sum_i f(x_i) = 1$

```python
import math

# Given three values
v1 = 1.0
v2 = 2.0
v3 = 3.0

# compute softmax
total = math.exp(v1) + math.exp(v2) + math.exp(v3)

s1 = math.exp(v1)/total
s2 = math.exp(v2)/total
s3 = math.exp(v3)/total

# print out
print(f"{s1:.5f}  {s2:.5f}   {s3:.5f}")
```

**Stable Softmax function**

$m = max(x)$

$f(x_i) = \frac{e^{x_i - m}}{\sum_j e^{x_j - m}}$



```python
import math

# Given three values
v1 = 1.0
v2 = 2.0
v3 = 3.0

# get max
max_value = v3

# compute stable softmax
e_v1 = math.exp(v1 - max_value)
e_v2 = math.exp(v2 - max_value)
e_v3 = math.exp(v3 - max_value)

total = e_v1 + e_v2 + e_v3

s1 = e_v1/total
s2 = e_v2/total
s3 = e_v3/total

# print out
print(f"{s1:.5f}  {s2:.5f}   {s3:.5f}") 
```
